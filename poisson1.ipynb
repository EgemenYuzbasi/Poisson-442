```python
%matplotlib notebook
import matplotlib.pyplot as plt
import scipy.odr.odrpack as odrpack
from scipy.optimize import curve_fit
import numpy as np

x=np.array([2.4761488507989844,3.2442645584558183,7.803780487622002,10.328444591358744])

y=np.array([2.4603788188512374,3.24533366432431,7.965195692137099,11.472257519005993])


def f(p, x):
    B,c=p
   
    return B*x+c


linear = odrpack.Model(f)
# mydata = odrpack.Data(x, y, wd=1./np.power(sx,2), we=1./np.power(sy,2))
mydata = odrpack.RealData(x, y)

myodr = odrpack.ODR(mydata, linear, beta0=[0.,1.])

myoutput = myodr.run()

myoutput.pprint()


x_fit = np.linspace(x[0], x[-1], 100)


y_fit = f(myoutput.beta, x_fit)

plt.plot(x_fit,y_fit, label='ODR', lw=3, color='purple')

plt.xlabel('Root of Mean')
plt.ylabel('Sigma')
plt.title('Standard Deviation vs Root of Mean')

plt.scatter(x,y)
plt.show()

plt.savefig('fit')
```
